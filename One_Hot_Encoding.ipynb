Using the data file found [here](https://https://raw.githubusercontent.com/nurfnick/Data_Viz/main/AB_NYC_2019.csv) on [ AirB&B in NYC](https://www.kaggle.com/dgomonov/new-york-city-airbnb-open-data).

Importing pandas library.

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/nurfnick/Data_Viz/main/AB_NYC_2019.csv")
df
```

Pandas pd.get_dummies() will turn your categorical column (neighbourhood_group & room_type) into indicator columns.

```python
pd.get_dummies(df.neighbourhood_group)
```

1 represents true and 0 represents false.

```python
pd.get_dummies(df.room_type)
```

```python
df.groupby('room_type').host_id.agg('count')
```

```python
neighbour = sorted(df.neighbourhood_group.unique())
neighbour 
```

```python
edd = {}

for i in range(5):
  edd[neighbour[i]] = i

edd
```

```python
df.groupby('neighbourhood_group').host_id.agg('count')
```

```python
df.neighbourhood_group.map(edd)
```

```python
df.neighbourhood_group=df.neighbourhood_group.map(edd)
df
```
