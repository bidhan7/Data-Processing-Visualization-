```python
import pandas as pa

df = pa.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/AB_NYC_2019.csv')

df.head()
```

```python
df.isnull().sum(axis = 0)
```

So, we got 4 collums which have null values. in which last review and reviews are highest one.
lets work in review per month.

```python
df.reviews_per_month = df.reviews_per_month.fillna(value = 0)
df
```

```python
df.last_review = df.last_review.fillna(value = 0)
df
```
