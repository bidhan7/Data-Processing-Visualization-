```python
!apt update
!apt install chromium-chromedriver
!pip install selenium
# set options to be headless
from selenium import webdriver
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
# open it, go to a website, and get results
driver = webdriver.Chrome('chromedriver',options=options)
```

Now with all that done, I am going to use selenium commands to crawl through the webpage and get the work done.

I'm importing` By `and `Keys` from selenium to get to a searcbox and type some inputs.

I first get the xpath of a google search box UI and use send_keys method to type a keyword and press enter.

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys 

url = "https://www.google.com"
driver.get(url)
element = driver.find_element(By.XPATH, '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input')
element.send_keys('Pizza')
element.send_keys(Keys.ENTER)
```

After this code, I get a google search page which searches about the keyword provided above. In this case, I get a result of a superbowl keyword.

```python
driver.current_url
```

Now, I am going to navigate to the first result and open the link associated with it. That means, I am opening a first search result from a google search.

```python
result = driver.find_element(By.CLASS_NAME, 'yuRUbf')
result.click()
```

Since the first google search result is of nfl official website so I am going to call current_url method to get where we are at which should match nfl.com.

```python
print(driver.current_url)
driver.title
```

I need to return the html of this page so lets get to work with beautiful soup and html parser.

```python
soup = BeautifulSoup(driver.page_source)
# print the html from a page
soup
```
