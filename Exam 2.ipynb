# (A)
Storing the data as a pandas in df and examining each datatype 

```python
df = pd.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Activity_Dataset_V1.csv')
df
```

```python
df.dtypes
```

# (B)
Removing the column that repeats the indexes and is ‘unnamed’ as a column. Using del method to delet that column.

```python
del df['Unnamed: 0']
df.head()
```

#(C) 
Pandas has an advantageous .str technique that you can use on text data. Since the column names are an 'index' type, you can utilize .str on them as well. 

```python
df.columns = df.columns.str.strip().str.replace('%', '').str.replace('(', '').str.replace(')', '')
df.head()
```

# (D)
Converting ‘activity day’ column into a datetime format.

```python
df['activity_day'] = pd.to_datetime(df['activity_day'])
df.activity_day.dtypes
```

# (E) 
Imputing ‘total steps’ by replacing the ‘NaN’s with an minimum number of steps. Using fillna method to replace. Converting float to int.

```python
df.total_steps.fillna(value=df.total_steps.min(), inplace=True)
df['total_steps'] = df['total_steps'].astype(int)
df
```

```python
df.dtypes
```

# (F)

First I convert avg_pace into datetime and then present it in float datatype.

```python
df['avg_pace'] = pd.to_datetime(df['avg_pace'])
df['avg_pace'] = df['avg_pace'].dt.hour + (df['avg_pace'].dt.minute)/60
df['avg_pace'].head(10)

```

# (G)
 I will groupby ‘workout type’ and find the mean, median, count and standard deviation of ‘calories’.

```python
df.groupby('workout_type').calories.agg(['mean','median','count', 'std'])
```

# (H)
I am making a indicator column that identifies if the activity achieved 30% or more ‘aerobic’ activity.

```python
df['indicator'] = np.where(df['aerobic'] >= (30), 'True', 'False')
df.head(15)
```

# (I)
Pandas DataFrame.loc attribute access a group of rows and columns by label or a boolean array in the given DataFrame. 

```python
d = df.loc[df.max_cadence == df.max_cadence.max()]
print('Day of Week = ' + d.activity_day.dt.day_name().iloc[0])
print('Workout type = ' + d.workout_type.iloc[0])
print('Maximum cadence = ' + str(df['max_cadence'].max()))
```
