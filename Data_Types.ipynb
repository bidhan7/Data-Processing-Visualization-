# Data Types

### Pandas

Pandas is a Python library. Pandas is used to analyze data. It can handle csv, excel, html and JSON from an sql call.

```python
import pandas as pa

df = pa.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/bank.csv')

df.head()
```

```python
df.age.dtypes
```

The first column has data type of 'int64' value. I think integer is best data type for the first column as age is usually not expressed in decimals.

The second column has data type shown 'O' (represents object), which means it has string data type.

```python
df.job.dtypes
```

```python
df.job = df.job.astype('category')

df.dtypes
```


I don't think it is an appropriate idea to assign an order to second column job because job dtypes doesn't have natural order. I would say job will be nominal categorical variable in this table.

But we can assign an order the the fourth column education. As we all know primary comes first and secondary and tertiary. Value seems to be ordinal. So it will be appropriate to make them in order. I will try to order them below:

```python
from pandas.api.types import CategoricalDtype

education_type = CategoricalDtype(categories=['primary', 'secondary', "tertiary"], ordered=True)

df_education = df.education.astype(education_type)
df
```
